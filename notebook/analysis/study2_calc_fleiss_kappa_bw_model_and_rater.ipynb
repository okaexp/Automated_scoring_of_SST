#### 以下、answer_and_rubricの結果でFleiss kappaを求める

```r
dir_answer_and_rubric = '/content/drive/MyDrive/Colaboratory/Work/sst_auto_scoring/result/study2/study2_answer_and_rubric_train_and_eval/data_test.csv'
dat_answer_and_rubric = read.csv(dir_answer_and_rubric)
```

```r
head(dat_answer_and_rubric)
```

```r
cross_tab_dat_answer_and_rubric <- dat_answer_and_rubric %>% group_by(labels, pred) %>%
  tally %>%
  spread(labels, n)
```

```r
cross_tab_dat_answer_and_rubric
```

```r
#分類一致率の算出
dat_answer_and_rubric_concat_kappa <- dat_answer_and_rubric[,c("labels", "pred")]
kappa2(dat_answer_and_rubric_concat_kappa,weight = TRUE)
```
